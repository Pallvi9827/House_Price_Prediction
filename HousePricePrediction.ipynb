```python
import pandas as pd

# lead dataset from githhub repo

url = "https://raw.githubusercontent.com/Pallvi9827/House_Price_Prediction/refs/heads/main/data.csv"
data = pd.read_csv(url)

# display first 5 rows

data.head()
```

```python
print("Dataset Information : ")
print(data.info())

print("\nColumns names : ")
print(data.columns)

print("First 5 Rows : ")
print(data.head())

print("Basic Statistics : ")
print(data.describe())
```

```python
print("Missing values in each columns : ")
print(data.isnull().sum())
```

```python
# drop unnecessary columns

data_cleaned = data.drop(['date','street','city','statezip','country'],axis=1)

# display cleaned dataset structure
print("cleaned dataset columns : ")
print(data_cleaned.columns)
```

```python
#"Remove outliers in price greater than 2 million "
data_cleaned = data_cleaned[data_cleaned['price'] < 2000000]

# Remove unusual large houses

data_cleaned = data_cleaned[(data_cleaned['sqft_living'] <7000) & (data_cleaned['sqft_lot'] < 50000)]

#Display cleaned data's new shape
print("New dataset size : ", data_cleaned.shape)
```

```python
# save cleaned data as csv

data_cleaned.to_csv('house_Price_data_cleaned.csv',index = False)
```

```python

url='https://raw.githubusercontent.com/Pallvi9827/House_Price_Prediction/refs/heads/main/house_Price_data_cleaned.csv'
data= pd.read_csv(url)
data.head()
```

```python
# select features x and tanget values y

x= data[['bedrooms','bathrooms','sqft_living','floors','waterfront','view','condition','yr_built']] #features

y= data['price'] #target

#display the shape to confirm
print("Feature's shape : ",x.shape)
print("Tanget's shape : ",y.shape)
```

```python
# splitting the model for training and testing

from sklearn.model_selection import train_test_split

# split the data
x_train ,x_test ,y_train ,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

#display the size of each set

print(" Training features shape : ", x_train.shape)
print(" Testing features shape : ", x_test.shape)
print(" Training target shape : ",y_train.shape)
print(" Testing target shape : ", y_test.shape)
```

```python
# Train Model

from sklearn.linear_model import LinearRegression

# Initialize the model
model = LinearRegression()

model.fit(x_train,y_train)

print("Model training complete")
```

```python
#Test Model

# make predictions
y_pred = model.predict(x_test)

# Display some predictions alongside with actual values

comparison = pd.DataFrame({'Actual':y_test.values,'Predicted':y_pred})
print(comparison.head(10))
```

```python
from sklearn.metrics import mean_absolute_error , mean_squared_error , r2_score
import numpy as np

mae = mean_absolute_error(y_test,y_pred)
mse = mean_squared_error(y_test,y_pred)
rmse= np.sqrt(mse)
r2 = r2_score(y_test,y_pred)

print(f"mean absolute error : {mae}")
print(f"mean squarred error : {mse}")
print(f"Root mean squarred error {rmse}")
print(f"R2 score :",{r2})
```
